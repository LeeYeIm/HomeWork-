```
#1. python 기본 문법

a=int(input())
def multi(a):
  i=1
  while i<10:
    print(a*i)
    i=i+1
   

%timeit multi(a)   
  
```

```
#2. numpy를 사용한 방법
import numpy as np

a=int(input())
def multi(a):
  x1=np.arange(1,10)
  
  x2=a*x1
  
  print(*x2, sep='\n')

%timeit multi(a)
```

```
#3. numba.jit을 이용한 방법

import numba

@numba.jit("void(f4[:])")


def multi(a):
  i=1
  while i<10:
    print(a*i, sep='\n')
    i=i+1

a=int(input())

%timeit multi(a)   
  
```

```
#4. cuda를 이용한 방법

import numpy as np
from numba import vectorize
import os

os.environ['NUMBAPRO_LIBDEVICE'] = "/usr/local/cuda-10.0/nvvm/libdevice"
os.environ['NUMBAPRO_NVVM'] = "/usr/local/cuda-10.0/nvvm/lib64/libnvvm.so"

@vectorize([int32(int32, int32)], target='cuda')

def multi(a,x1):
  x2=a*x1
  
  return x2
a=int32(input())
x1=np.arange(1,10, dtype=np.int32)

%timeit print(*multi(a,x1), sep='\n') 
  
```
